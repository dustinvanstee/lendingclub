```python
# This function builds on my experiment above to fetch lendingclub data and return a dict of k/v pairs.  
def get_data(url) :
    rv = {}
    test = browser.open(url)
    soup = BeautifulSoup(test.text, features="lxml")
    tables = soup.findAll('table')
    if(len(tables) == 0) :
        print("Need to re-authenticate !")
    else : 
        mytable = tables[1]
        rows = mytable.find_all('tr')

        for r in rows :
            key = r.find('th').text
            value = r.find('td').text
            rv[key] = value
    
    return rv
```

```python
import json
def save_data(input_data) :
    #print(input_data)
    f = open("test.json", 'w')
    f.write(json.dumps(input_data))
    f.close()

def load_data() :
    f = open('test.json', 'r')
    raw = f.read()
    json_data = json.loads(raw)
    return json_data
             
#save_data(data_list)
new_list = load_data()            
```

```python
#loan_df.describe()
# use loan id, as a dictionary input

#data_list = ["None"] * loan_id_url_df.count()[1]

# load_data(data_list)

def get_more_data(max_idx) :
    authenticate()
    data_list = load_data()
    i=0
    print("Starting max_idx = {}".format(max_idx))
    for index, row in loan_id_url_df.iterrows():
        
        if(data_list[i] == 'None' ):
            #print("i={} id={} url={}".format(index,row["id"],row["url"]))
            try :
                loan_data = get_data(row["url"])
            except :
                print("Record {} unavailable.  Saving as blank")
                data_list[i] = [i,row["id"],row["url"], 'na','na',
                    'na', 'na']
            else :
                try :
                    data_list[i] = [i,row["id"],row["url"], loan_data['Note Status'],loan_data['Note issued on'],
                        loan_data['Funding Received'], loan_data['Investors']]
                except :
                    print("loan data does not exist .  this is what i got {}:{}".format(i, loan_data))
        
        if(i > max_idx ):
            break
        i+=1
    print("Data through idx {} captured!".format(max_idx))
    save_data(data_list)

for idx in range(50000,100000,1000):
    get_more_data(idx)
    
```
